```python
import networkx as nx
from networkx.drawing.nx_agraph import graphviz_layout
import matplotlib.pyplot as plt

## whenever registering a child, adding an edge to the tree
tree = {
    (0, (), 12): [90, 1],
        (1, (0), 11): [80, 1],
            (2, (0, 1), 10): [80, 1],
        (2, (0), 11.5): [75, 0.5],
            (1, (0, 2), 10): [20, 0.5],
}
edge_labels={
    (0, (0,1)):[90,1],
        ((0,1), (0,1,2)):[80, 1],
    (0, (0,2)):[75, 0.5],
        ((0,2), (0,2,1)):[20, 0.5],
}

dG = nx.DiGraph()
labels = {}
pos = {}

node_size = 200
font_size = 10
edge_width = 3
arrow_size = 20

for k, v in edge_labels.items():
    nfrom, nto = k
    dG.add_edge(nfrom, nto)
for node in dG.nodes():
    labels[node] = str(node)
    
pos=graphviz_layout(dG, prog='dot')
plt.box(False)
nx.draw_networkx_nodes(dG, pos, node_color='#F2F09E', node_shape='s', node_size=node_size, alpha=1)
# nx.draw_networkx_nodes(dG, pos, node_color='#1abc9c', node_shape='s', nodelist=[current_node], node_size=node_size, alpha=1)
# nx.draw_networkx_nodes(dG, pos, node_color='#f09289', node_shape='s', nodelist=[max_node], node_size=node_size, alpha=1)
nx.draw_networkx_labels(dG, pos, labels, font_size=font_size)
nx.draw_networkx_edges(dG, pos, edge_color='#F2F09E', width=edge_width, arrowsize=arrow_size, alpha=1)
nx.draw_networkx_edge_labels(dG, pos, edge_labels=edge_labels, font_size=font_size)
# plt.set_size_inches(size['inchwidth'], size['inchheight'])
plt.show()
```
