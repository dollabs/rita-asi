```python
import csv

with open('../maps/6by6_1.csv', encoding='utf-8-sig') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    world_grids = []
    for row in csv_reader:
        world_grids.append(row)
    world_height = len(world_grids)
    world_width = len(world_grids[0])

tile_index = 0
tile_indices = {} # mapping (x,y) coord to tile index
tilesummary = {}
reward_by_grid = {'':0, 'S':0, 'G': 10}
for i in range(world_height):
    for j in range(world_width):
        grid = world_grids[i][j]
        if grid == 'S':
            start_index = tile_index
        if grid != 'W':
            grid_states = {}
            grid_states['pos'] = (i,j)
            tile_indices[(i,j)] = tile_index
            grid_states['reward'] = reward_by_grid[grid]
            if j == world_width - 1 or world_grids[i][j+1] == 'W':
                grid_states[0] = 'wall'
            if i == 0 or world_grids[i-1][j] == 'W':
                grid_states[90] = 'wall'
            if j == 0 or world_grids[i][j-1] == 'W':
                grid_states[180] = 'wall'
            if i == world_height - 1 or world_grids[i+1][j] == 'W':
                grid_states[270] = 'wall'
            tilesummary[tile_index] = grid_states
            tile_index += 1

for k in tilesummary.keys():
    tile = tilesummary[k]
    i,j = tile['pos']
    if 0 not in tile.keys():
        tile[0] = tile_indices[(i,j+1)]
    if 90 not in tile.keys():
        tile[90] = tile_indices[(i-1,j)]
    if 180 not in tile.keys():
        tile[180] = tile_indices[(i,j-1)]
    if 270 not in tile.keys():
        tile[270] = tile_indices[(i+1,j)]
    print(tilesummary[k])
    
print(start_index)
```

```python
def diff(light,shadow):
    new = pd.DataFrame('', index=range(len(light)), columns=range(len(light[0])))
    for i in range(len(light)):
        for j in range(len(light[0])):
            if light[i][j] == 1 and shadow[i][j] != 'W':
                new[i][j] = '*'
            if light[i][j] == 'W':
                new[i][j] = 'W'
    return new
    
def print_shadow(grids, pos, head):
    y = pos[0]
    x = pos[1]
    light = pd.DataFrame('', index=range(world_height), columns=range(world_width))
    shadow = pd.DataFrame('', index=range(world_height), columns=range(world_width))
    for k in tilesummary.keys():
        tile = tilesummary[k]
        j,i = tile['pos']
        light[i][j] = 0
        if j-y <= 3 and j-y >= -3 and i-x <= 3 and i-x >= -3:
            if head == 0:
                if - (i-x) <= j-y and j-y <= (i-x):
                    light[i][j] = 1
            if head == 90:
                if j-y <= (i-x) and j-y <= -(i-x):
                    light[i][j] = 1
            if head == 180:
                if (i-x) <= j-y and j-y <= - (i-x):
                    light[i][j] = 1
            if head == 270:
                if j-y >= (i-x) and j-y >= -(i-x):
                    light[i][j] = 1
        
    for i in range(world_width): # x axis
        for j in range(world_height): # y axis
            if light[i][j] == '':
                light[i][j] = 'W'
                
                if head == 0 and i>x:
                    for k in range(i,world_width):
                        shadow[k][j] = 'W' # the same row
                    
                    if y<j:
                        if i+1 < world_width and j+1 < world_height:
                            for k in range(i+1,world_width): # the row + 1
                                shadow[k][j+1] = 'W'
                        if i+2 < world_width and j+2 < world_height:
                            for k in range(i+2,world_width): # the row + 2
                                shadow[k][j+2] = 'W'
                    if y>j:
                        if i+1 < world_width and j-1 >=0:
                            for k in range(i+1,world_width): # the row - 1
                                shadow[k][j-1] = 'W'
                        if i+2 < world_width and j-2 >=0:
                            for k in range(i+2,world_width): # the row - 2
                                shadow[k][j-2] = 'W'
                            
                if head == 90 and j<y:
                    for k in range(0,j+1):
                        shadow[i][k] = 'W'
                        
                    if i>x:
                        if i+1 < world_width and j+1 >= 0:
                            for k in range(0,j+1): # the col + 1 
                                shadow[i+1][k] = 'W'
                        if i+2 < world_width and j+2 >= 0:
                            for k in range(0,j+2): # the col + 2
                                shadow[i+2][k] = 'W'
                    if i<x:
                        if i-1 >= 0 and j+1 >= 0:
                            for k in range(0, j+1): # the col - 1 
                                shadow[i-1][k] = 'W'
                        if i-2 >= 0 and j+2 >= 0:
                            for k in range(0, j+2): # the col - 2
                                shadow[i-2][k] = 'W'
                    
                if head == 180 and i<x:
                    for k in range(0,i+1):
                        shadow[k][j] = 'W'
                     
                    if y<j:
                        if i+1 < world_width and j+1 < world_height:
                            for k in range(0, i+1): # the row + 1
                                shadow[k][j+1] = 'W'
                        if i+2 < world_width and j+2 < world_height:
                            for k in range(0, i+2): # the row + 2
                                shadow[k][j+2] = 'W'
                    if y>j:
                        if i+1 < world_width and j-1 >=0:
                            for k in range(0, i+1): # the row - 1
                                shadow[k][j-1] = 'W'
                        if i+2 < world_width and j-2 >=0:
                            for k in range(0, i+2): # the row - 2
                                shadow[k][j-2] = 'W'
                                
                if head == 270 and j>y:
                    for k in range(j,world_height):
                        shadow[i][k] = 'W'
                        
                    if i>x:
                        if i+1 < world_width and j+1 < world_height:
                            for k in range(j+1,world_height): # the col + 1 
                                shadow[i+1][k] = 'W'
                        if i+2 < world_width and j+2 < world_height:
                            for k in range(j+2,world_height): # the col + 2
                                shadow[i+2][k] = 'W'
                    if i<x:
                        if i-1 >= 0 and j+1 < world_height:
                            for k in range(j+1,world_height): # the col - 1 
                                shadow[i-1][k] = 'W'
                        if i-2 >= 0 and j+2 < world_height:
                            for k in range(j+2,world_height): # the col - 2
                                shadow[i-2][k] = 'W'
#     display(light)
#     display(shadow)
    display(diff(light,shadow))
    return diff(light,shadow)

# to check all tiles
if False:
    angles = [180]#,[0, 90, 180, 270]
    for k in tilesummary.keys():
        tile = tilesummary[k]
        i,j = tile['pos']
        for k in range(len(angles)):
            print(i,j, angles[k])
            print_shadow(grids, (i,j), angles[k])
    
print_shadow(grids, (5,0), 0)
```

```python
import string
digs = string.digits + string.ascii_letters

def int2base(x, base):
    if x < 0:
        sign = -1
    elif x == 0:
        return digs[0]
    else:
        sign = 1

    x *= sign
    digits = []

    while x:
        digits.append(digs[int(x % base)])
        x = int(x / base)

    if sign < 0:
        digits.append('-')

    digits.reverse()

    return ''.join(digits)

int2base(4, 2)
```
