```python
import time
import turtle
from turtle import Turtle

screen = turtle.Screen()
screen.setup(width = 700, height = 700) # window size
screen.tracer(0,0)

note_time = Turtle()
note_time.hideturtle()
note_time.up()

img = "../texture/12/fire.gif"
screen.register_shape(img)
note_time.shape(img)

index = 0
while True:
    start = time.time()
    
    note_time.clear()
    note_time.goto(-330, -250)
    note_time.write(str(index), font=("Courier", 14, 'normal'))
    print(round(time.time()-start, 3))
```
